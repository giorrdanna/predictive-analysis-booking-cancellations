Prever uma variavel continua - regressão linear
Todas variaveis têm de estar convertidas em números


Passos para aplicar a Regressão Linear
Definir a variável dependente (target) e as variáveis independentes (features):

A regressão linear requer que você defina uma variável dependente contínua que deseja prever.
No seu caso, você pode querer prever o Billing Amount (ou outra variável contínua).
Pré-processamento dos dados:

Certifique-se de que todas as variáveis categóricas foram devidamente convertidas para numéricas.
Verifique a normalidade e multicolinearidade entre as variáveis. Se necessário, remova variáveis altamente correlacionadas.
Separar o dataset em treinamento e teste:

Divida o dataset em conjunto de treinamento e teste usando train_test_split.
Treinar o modelo de Regressão Linear:

Aplique a regressão linear usando LinearRegression do Scikit-Learn.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Carregar o dataset utilizando o caminho correto
file_path = r"hotel_booking.csv"
df = pd.read_csv(file_path)

# Remover as linhas com valores ausentes
df_clean = df.dropna()

# Definir as variáveis independentes (features) e dependente (target)
X_clean = df_clean.select_dtypes(include=['float64', 'int64']).drop('adr', axis=1)  # Features: todas as colunas numéricas menos 'adr'
y_clean = df_clean['adr']  # Target: 'adr' (Average Daily Rate)

# Dividir o dataset em treino (70%) e teste (30%)
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.3, random_state=42)

# Instanciar o modelo de regressão linear
model = LinearRegression()

# Treinar o modelo
model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = model.predict(X_test)

# Avaliar o modelo
r2 = r2_score(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # Raiz do erro quadrático médio
mae = mean_absolute_error(y_test, y_pred)

# Exibir os resultados
print(f"R²: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
```

Regressão logistica - Prever variável discreta


A regressão logística é um modelo de classificação adequado para prever variáveis categóricas, especialmente quando há apenas duas classes (classificação binária). No entanto, você também pode usá-la para problemas com múltiplas classes, como o seu caso, que envolve as categorias Normal, Abnormal, e Inconclusive na variável Test Results.

Como aplicar Regressão Logística ao seu dataset
Aqui estão os passos:

Selecionar as Features e a Variável Alvo (Target):

Variável alvo: Test Results (categórica).
Features: todas as outras variáveis que você considera úteis para prever o resultado dos testes.
Divisão dos Dados:

Divida os dados em conjunto de treino e conjunto de teste.
Treinamento do Modelo:

Use o modelo de Regressão Logística Multiclasse (padrão multinomial).
Avaliação do Modelo:

Métricas como Acurácia, Precision, Recall, F1-Score e Matriz de Confusão podem ser usadas para avaliar o desempenho.


Explicação dos parâmetros:
multi_class='multinomial': Especifica que estamos lidando com múltiplas classes (mais de duas).
solver='lbfgs': Otimizador eficiente para problemas multiclasse.
max_iter=1000: Número máximo de iterações para garantir a convergência.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Separar as features (X) e a variável alvo (y)
X = df1.drop('is_canceled', axis=1)  # Features: todas as variáveis exceto 'is_canceled'
y = df1['is_canceled']  # Target: 'Test Results'

# Dividir o dataset em treino (70%) e teste (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Instanciar o modelo de Regressão Logística
model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, random_state=42)

# Treinar o modelo
model.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = model.predict(X_test)

# Avaliar o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia: {accuracy:.4f}")

# Relatório detalhado de classificação
print("Relatório de Classificação:")
print(classification_report(y_test, y_pred))

# Matriz de Confusão
print("Matriz de Confusão:")
print(confusion_matrix(y_test, y_pred))
```

knn - Discreta

O K-Nearest Neighbors (KNN) é outro algoritmo de classificação que também pode ser aplicado ao seu problema de previsão dos resultados de teste. O KNN é um algoritmo simples e intuitivo que classifica um ponto de dados com base nos "K" vizinhos mais próximos no espaço de características. Ele pode ser usado tanto para classificação binária quanto para classificação multiclasse, como no seu caso.

Passos para aplicar o KNN ao seu dataset:
Selecionar as Features e a Variável Alvo (Target):

Variável alvo: Test Results (categórica).
Features: as outras variáveis numéricas que ajudarão a prever o resultado.
Padronização dos Dados:

O KNN é sensível às escalas das features. Por isso, é importante padronizar os dados (colocar em uma escala comum).
Divisão dos Dados:

Dividir o dataset em conjunto de treino e conjunto de teste.
Treinamento do Modelo:

Definir o valor de K (o número de vizinhos a considerar) e ajustar o modelo.
Avaliação do Modelo:

Usar métricas como Acurácia, Matriz de Confusão, e F1-Score para avaliar o desempenho.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Separar as features (X) e a variável alvo (y)
X = df1.drop('is_canceled', axis=1)  # Features: todas as variáveis exceto 'is_canceled'
y = df1['is_canceled']  # Target: 'is_canceled'

# Dividir o dataset em treino (70%) e teste (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Padronizar as features para KNN (normalizar os dados)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Instanciar o modelo KNN com K=5 (você pode ajustar esse valor)
knn = KNeighborsClassifier(n_neighbors=5)

# Treinar o modelo
knn.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = knn.predict(X_test)

# Avaliar o modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Acurácia: {accuracy:.4f}")

# Relatório detalhado de classificação
print("Relatório de Classificação:")
print(classification_report(y_test, y_pred))

# Matriz de Confusão
print("Matriz de Confusão:")
print(confusion_matrix(y_test, y_pred))

```

```python
import pandas as pd
from scipy.stats import shapiro

# Carregar os dados do CSV em um DataFrame
df = pd.read_csv('hotel_booking.csv')

# Contar o número de missings
missing_counts = df.isnull().sum()

# Filtrar apenas as variáveis com missings
missing_vars = missing_counts[missing_counts > 0]

# Exibir as variáveis com missings
print("Variáveis com Missings:")
print(missing_vars)

# Analisar o tipo de cada variável e sua distribuição
for column in missing_vars.index:
    if df[column].dtype in ['int64', 'float64']:
        # Teste de normalidade
        stat, p = shapiro(df[column].dropna())
        if p > 0.05:
            distribution = 'normal'
        else:
            distribution = 'não normal'
        print(f"{column}: {df[column].dtype}, {distribution}")
    else:
        print(f"{column}: {df[column].dtype}, categórica")
```

```python
# Substituir os missings
for column in missing_vars.index:
    if df[column].dtype in ['int64', 'float64']:
        # Teste de normalidade
        stat, p = shapiro(df[column].dropna())
        if p > 0.05:
            # Distribuição normal
            mean_value = df[column].mean()
            df[column].fillna(mean_value, inplace=True)
        else:
            # Distribuição não normal
            median_value = df[column].median()
            df[column].fillna(median_value, inplace=True)
    else:
        # Variável categórica
        mode_value = df[column].mode()[0]
        df[column].fillna(mode_value, inplace=True)

# Verificar se ainda há missings
print("Número de missings após o tratamento:")
print(df.isnull().sum())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE

# Carregar os dados
df1 = pd.read_csv('hotel_booking.csv')

# Filtrar as colunas necessárias
X_filtered = df1[['lead_time', 'adr']]
y = df1['is_canceled']

# Dividir o dataset em train e test
X_train, X_test, y_train, y_test = train_test_split(X_filtered, y, test_size=0.25, random_state=42)

# Instanciar o SMOTE
smote = SMOTE(random_state=42)

# Número de iterações para repetição
n_iterations = 10

# Lista para armazenar os modelos
models = [LogisticRegression(max_iter=1000), RandomForestClassifier()]

# Função para treinar e avaliar
def train_and_evaluate(models, X_train, y_train, X_test, y_test):
    for model in models:
        model.fit(X_train, y_train)
        score = model.score(X_test, y_test)
        print(f'Modelo: {model.__class__.__name__}, Acurácia: {score:.4f}')

# Repetir amostras do train e obter modelos
for i in range(n_iterations):
    # Amostragem com SMOTE
    X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

    # Treinar e avaliar
    train_and_evaluate(models, X_resampled, y_resampled, X_test, y_test)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar apenas colunas necessárias
df1 = pd.read_csv('hotel_booking.csv', usecols=['lead_time', 'adr', 'is_canceled'])

# Reduzir tipos de dados
df1['lead_time'] = df1['lead_time'].astype('int32')
df1['adr'] = df1['adr'].astype('float32')

# Remover colunas com muitos valores nulos
df1.dropna(axis=1, inplace=True)

# Calcular a matriz de correlação
correlation_matrix = df1.corr()

# Plotar o heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap de Correlação')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Filtrar as colunas para features e target
X = df1[['lead_time', 'adr']]  # Features
y = df1['is_canceled']  # Target

# Dividir o dataset em train (75%) e test (25%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Modelos a serem usados
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Naive Bayes': GaussianNB()
}

# Treinar e avaliar cada modelo
for model_name, model in models.items():
    # Treinamento
    model.fit(X_train, y_train)

    # Previsão
    y_pred = model.predict(X_test)

    # Avaliação
    print(f'\n{model_name}')
    print(f'Acurácia: {accuracy_score(y_test, y_pred):.4f}')
    print('Relatório de Classificação:')
    print(classification_report(y_test, y_pred))
```
